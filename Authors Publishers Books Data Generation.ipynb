#### Replace Author names with their IDs

```python
# Create a dictionary mapping authors to their IDs
authid_map = dict(zip(The_Authors['Author_Name'], The_Authors['AuthorID']))

# Replace author names in the books dataframe with their corresponding IDs
books['Author_ID'] = books['Book-Author'].map(authid_map)
```

```python
books.isnull().sum()
```

```python
books.shape
```

#### Replace publisher names with their IDs

```python
# Create a dictionary mapping authors to their IDs
pubid_map = dict(zip(The_Publisher['Publisher_Name'], The_Publisher['PublisherID']))

# Replace author names in the books dataframe with their corresponding IDs
books['Publisher_ID'] = books['Publisher'].map(pubid_map)
```
